```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image_sudoku = cv2.imread('pic/sudoku.tif')
plt.imshow(image_sudoku,cmap = 'gray')
image_sudoku = cv2.cvtColor(image_sudoku,cv2.COLOR_BGR2GRAY)
```

```python
histogram_1 = cv2.calcHist([image_sudoku],[0],None,[256],[0,256])
plt.plot(histogram_1)
plt.show()
```

```python
equalize_image = cv2.equalizeHist(image_sudoku) 
plt.imshow(equalize_image , cmap = 'gray')
# cv2.imwrite('image1.jpg',equalize_image)
```

```python
histogram_2 = cv2.calcHist([equalize_image],[0],None,[256],[0,256])
plt.plot(histogram_2)
plt.show()
```

```python
print(equalize_image.shape)
rows , columns = equalize_image.shape
```

```python
images = []
for image in range(1,8):
    img = cv2.imread(f'pic/result/img{image}.jpg')
    img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    img = cv2.resize(img,(400,400))
    images.append(img)
```

```python
image_result1 = cv2.hconcat([images[0],images[1]])
image_result2 = cv2.hconcat([images[2],images[3]])
images1 = cv2.vconcat([image_result1,image_result2])

image_result3 = cv2.hconcat([images[4],images[5]])
image_result4 = cv2.hconcat([images[6],images[2]])
images2 = cv2.vconcat([image_result3,image_result4])

images_result = cv2.vconcat([images1,images2])

plt.imshow(images_result,cmap = 'gray')
# cv2.imwrite('image2.jpg',images_result)
```
